```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dates = pd.date_range('20130101', periods=8)
print(dates)
s1 = pd.Series([1,2,3,4,5,6,7,8], index=pd.date_range('20130102', periods=8))

df = pd.DataFrame(np.random.randn(8,4), index=dates, columns=list('ABCD'))

#add s1 a df
df['F'] = s1
#asigno a a en 0  valor 0
df.at[dates[0],'A']=0
print('[--]')
print(df)
#al i[0]j[1]le asigno 0 
df.iat[0,1]=0
print('[-------]')
print(df)
#a la columna d le asigno 5 a toda la profundidad del df en la columna D
df.loc[:,'D']=np.array([5]*len(df))
print('[-------]')
print(df)
# POner a negativo todo lo que sea mayor que 0
df2=df.copy()
df2[df2>0]=-df2

print('[-------]')
print(df2)
# asigno un new data frama con los 4 primeros elementos de la columna
# dates de 0a 4 y adiciono una columna E vacia
df1=df.reindex(index=dates[0:4],columns=list(df.columns)+['E'])
print('[-------]')
print(df1)
# 
print(df1.columns)
#asigno los valores de 1 a las 2 primeras
# filas donde las fechas sean 
# iguales  a las de 0 a 1 de dates
df1.loc[dates[0]:dates[1],'E']=1

print('[-------]')
print(df1)

#eleminar  todas las filas que tengas valor nan
df1.dropna()
print('[-------filas que tengas valor nan]')
print(df1)
#rellenar los elementos que no tengan valor 
df1.fillna(4)
print('[-------]')
print(df1)

# muestra una matriz de boolean donde sean nan o no
df1.isna()
print('[-------]')
print(df1)
#promedio de la matriz total ordenada por columnas
print('[-------mean]')
print(df1.mean())

#promedio por filas 
print('[-------mean 1]')
print(df1.mean(1))

s = pd.Series([1,3,5,np.nan,6,8,12,12], index=dates).shift(2)
print('[-------new s]')
print (s)

#uniando doa matrices
print('[-------new s un]')
print(df.sub(s,axis='index'))

#aplicando funciones a los datos
print('[------- np.cumsum eje x]')
print(df.cumsum(axis=1))

print('[------- np.cumsum eje y]')
print(df.cumsum(axis=0))

#

print('[------- aply lamdas maximo menos minimo de cada columna')
df.apply(lambda x: x.max() - x.min())

#histogramas 
#creando matriz con ramdom

s= pd.Series(np.random.randint(0,7,size=10))
print(s)

#contar los valores de una matriz
s.value_counts()


```
